Stop words are commonly used, high-frequency words—such as "the", "is", "and", "in"—that are typically excluded during natural language processing (NLP) and text analysis. They are removed to reduce noise and computational load, allowing models to focus on words with more unique semantic meaning, thus improving the efficiency of tasks like sentiment analysis and information retrieval

```python
paragraph = """I have three visions for India. In 3000 years of our history people from all over the world have come and invaded us, captured our lands, conquered our minds. From Alexander onwards the Greeks, the Turks, the Moguls, the Portuguese, the British, the French, the Dutch, all of them came and looted us, took over what was ours. Yet we have not done this to any other nation. We have not conquered anyone. We have not grabbed their land, their culture and their history and tried to enforce our way of life on them. Why? Because we respect the freedom of others. That is why my FIRST VISION is that of FREEDOM. I believe that India got its first vision of this in 1857, when we started the war of Independence. It is this freedom that we must protect and nurture and build on. If we are not free, no one will respect us.

We have 10 percent growth rate in most areas. Our poverty levels are falling. Our achievements are being globally recognised today. Yet we lack the self-confidence to see ourselves as a developed nation, self-reliant and self-assured. Isn’t this incorrect? MY SECOND VISION for India is DEVELOPMENT. For fifty years we have been a developing nation. It is time we see ourselves as a developed nation. We are among top five nations in the world in terms of GDP. We have a 10 percent growth rate in most areas. Our poverty levels are falling. Our achievements are being globally recognised today. Yet we lack the self-confidence to see ourselves as a developed nation, self-reliant and self-assured. Isn’t this incorrect? MY SECOND VISION for India is DEVELOPMENT. For fifty years we have been a developing nation. It is time we see ourselves as a developed nation. We have to take the confidence of our economic growth to the people, to every village, every home. We have to bridge the rural-urban divide. We have to develop our villages, our smalltowns. Only then will we have created a developed India. By 2020, I want"""
```

```python
from nltk.stem import PorterStemmer
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
#to get the list of stopwords in english
stopwords.words('english')
```

```python
#i will use stemming and stopword removal to preprocess the text data before applying any machine learning algorithms on it. Stemming will help me to reduce the words to their root form and stopword removal will help me to remove the common words that do not add much meaning to the text.
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
#tokenization is the process of breaking down a large paragraph into smaller units such as sentences or words. This is an important step in natural language processing as it allows us to analyze the text at a more granular level. In this case, I will use sentence tokenization to break down the paragraph into individual sentences.
sentences = nltk.sent_tokenize(paragraph)
```

```python
print(type(sentences))
```

```python
#Apply stopwords and filter then apply stemming to the sentences
for i in range(len(sentences)):
    #i need to get the words in the sentence and remove the stopwords and then apply stemming to the remaining words and then join them back to form the sentence
    words = nltk.word_tokenize(sentences[i])
    words = [word for word in words if word not in stopwords.words('english')]
    words = [stemmer.stem(word) for word in words]
    sentences[i] = ' '.join(words)
```
