Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes to the root of the words known as lemma. Stemming is important in natural language understandind and natural language processing

```python
## Classification Problem
## Comments of product is a positive review or a negative review
# for words like "eat", "eating", "eats" we will use the word "eat" as the root word and call it "lemma"
# we will use the "lemmatization" technique to find the root word of a word
```

```python
words = ["eat", "eating", "eaten","write", "writing", "writes","program", "programming", "programs","history","finally","finalize","finalized","congratulations","congratulate","congratulated","fairly","sportingly"]
```

##Porter Stemmer

```python
# we will use the "PorterStemmer" technique to find the root word of a word
from nltk.stem import PorterStemmer

#create a object of the PorterStemmer class
stemmer = PorterStemmer()
```

```python
for word in words:
    print(f"{word} --> {stemmer.stem(word)}")
```

RegexpStemmer Class

NLTK has Regexpstemmer class with the help of which we can easily implement Regular Expression Stemmer algorithims. It basically takes a single regular expression and removes a prefix or suffix that matches the expression.

```python
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$|ed$|ation$', min=4) # remove ing, s, e, able from the end of the word if the word is at least 4 characters long
```

```python
reg_stemmer.stem("eating")
```

```python
reg_stemmer.stem("ingprogram")
```

```python
for word in words:
    print(f"{word} --> {reg_stemmer.stem(word)}")
```

#Snowball Stemmer

```python
from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer("english") #it takes the language as an argument
```

```python
snowball_stemmer.stem("eating")
```

```python
for word in words:
    print(f"{word} ----> {snowball_stemmer.stem(word)}")
```

```python
for word in words:
    print(f"{word} --> {stemmer.stem(word)} --> {reg_stemmer.stem(word)} --> {snowball_stemmer.stem(word)}")
```
